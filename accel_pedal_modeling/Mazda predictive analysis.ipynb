# Explorations in Automotive data: Regeneration of the Accelerator Pedal Position (APP) parameter

In the automotive tuning world, the ability to tune an engine's calibration requires a good frequency of engine log data to make a decent calibration.  However, most stock engine computers in vehicles today are not made to output high rates of engine parameters; instead they are made to run the car off of models created from the factory.  For aftermarket tuners, this presents a problem as data points are sparce where the calibration sections are needed.  

In order to optimize the rate of data collected, many calibrators will turn off paremeter query items and attempt to operate in the absense of them.  The trade off occurs when trying to analyze a critical part of car's operation, called *wide open throttle*, or WOT.  During this part of the engine's operation, the monotonically increasing engine revolutions under load must be optimized correctly, or adverse drive behavior will occur.

While many calibrators attempt to operate in the absense of it, I felt that a better solution was possible via statistical and predictive methods.

**The goal for this solution is as follows**: *Determine the Accelerator Pedal Position (APP) parameter from the rest of the supplied log parameters, such that one may remove the APP log from the parameter list and obtain an increased logging rate from the engine computer. *

**Other details about the problem setup:**
- This data comes from a Mazda Mazdaspeed6 (MS6), from a leading calibrator in the industry
- The log list comes from a WOT pull
- The list of data is considered the "default" from the tuning industry for this vehicle

With the problem statement out of the way, let's focus on getting the analysis set up.

```python
import pandas as pd
import sklearn as sk
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
mz_file = pd.read_csv("Mazda 4th Gear Pull.csv")
```

The above code sets up the Python and data analysis tools for the problem.  This setup is relatively standard for the IPython Notebook community.  

I've read the file into a *Pandas* dataframe, such that data slicing and manipulation is easier.

Also for this instance, I know that the data's ranges vary wildly; I'll need to scale the data to be able to use some feature selection capabilities of Lasso.  We will re-run the tests with less scaled variables once we know what variables to use.

Lasso regression has some special properties in detecting collinearity and selection via shrinkage methods via the L1 norm. 

##Scaling of the data

```python
# Select the appropriate target data
target_app_scale = mz_file[mz_file.columns[1]]
# Get the rest of the feature data.
data_scale = mz_file.drop(mz_file.columns[[0,1]],1)

# Center the data
target_app_scale = (target_app_scale - target_app_scale.mean())/(target_app_scale.max() - target_app_scale.min())
data_scale = (data_scale - data_scale.mean())/(data_scale.max()-data_scale.min())
```

At this point, we'll start setting up the problem for Lasso variable selection.  We'll do a general Lambda sweep, and see what comes out of the result—we are looking for good model prediction with the least amount of dimensions of data.

Please note in this case it is normally best practice to do cross validation within Lasso, but in this case we are just doing an *intial* exploration.  Not doing cross validation risks overfitting the model, as an FYI.

##Running Lasso Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

lambda_array = [.0001, .0005, .001, .005, .01, .05, .1, .5, 1, 5]
coef_array = []
score_array = []
mse_array = []

for lam in lambda_array:
    clf = linear_model.Lasso(alpha=lam)
    clf.fit(data_scale, target_app_scale)
    coef_array.append(clf.coef_)
    
    score = clf.score(data_scale, target_app_scale)
    score_array.append(score)
    test_prediction = clf.predict(data_scale)
    mse_array.append(mean_squared_error(target_app_scale, test_prediction))
    
# Attach names back into the dataframe
final_coef = pd.DataFrame(coef_array, columns=data_scale.columns)
```

```python
plt.subplots(figsize=(12,6))
plt.title("Lasso Regression of features", fontsize=12)
plt.ylabel("Coefficients", fontsize=12)
plt.xlabel("-(Log)Lambda", fontsize=12)
plt.plot(-np.log10(lambda_array),final_coef)
plt.vlines(-np.log10(.001), -.2, .7, linestyle='--')
plt.show()

plt.subplots(figsize=(12,6))
plt.title("Lasso R^2 and MSE scores", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.xlabel("-(Log)Lambda", fontsize=12)
plt.plot(np.log10(lambda_array), score_array)
plt.plot(np.log10(lambda_array), mse_array)
plt.vlines(np.log10(.001), 0, 1, linestyle='--')
plt.show()
```

At this point, we'll select the 3rd lambda in run some tests with, which is lambda = .001.  The reason for this selection is a balance of availible variables and model score.

##Determining the necessary features for Linear Regression

```python
from sklearn import cross_validation

X_train, X_test, y_train, y_test = cross_validation.train_test_split(data_scale, target_app_scale, test_size=0.4, random_state=0)
```

```python
from sklearn.metrics import mean_squared_error
clf = linear_model.Lasso(alpha=.001)
clf.fit(X_train, y_train)
print "R^2 score: " + str(clf.score(X_test, y_test))
y_pred = clf.predict(X_test)
print "Mean Squared Error " + str(mean_squared_error(y_test, y_pred))
```

Using the scaled data, we then calculate for a static lambda Lasso test using lambda=.001.  We have a decent R^2 score, and MSE is relatively good.

Let's take a look at what features are left by re-attaching the dataframe.

```python
pd.DataFrame([clf.coef_], columns=data_scale.columns)
```

It looks like AFR, Throttle Position, Wastegate Duty, and Intake valve advance are the strongest predictors.  We will toss Intake Temp, since it does not have a very big coefficient left. 
  
In terms of scaling data, most of the data is relatively sane in scale, so we'll see what we can get without scaling on the lienar regression model.

Let's use standard least squares for the next prediction, along with cross validation sets.

##Running Linear Regression

```python
#Split the data
lm_target = mz_file[mz_file.columns[1]]
#Selecting the features of  AFR, Throttle Position, Wastegate Duty, and Intake valve advance 
lm_data = mz_file[mz_file.columns[[2,9,17,19]]]
#Split the data in to CV train and test
X_train, X_test, y_train, y_test = cross_validation.train_test_split(lm_data, lm_target, test_size=0.4, random_state=0)

clf = linear_model.LinearRegression(fit_intercept=False)
clf.fit(X_train, y_train)
print "R^2 score: " + str(clf.score(X_test, y_test))
y_pred = clf.predict(X_test)
print "Mean Squared Error " + str(mean_squared_error(y_test, y_pred))
```

```python
plt.plot(lm_target)
plt.plot(clf.predict(lm_data))
plt.show()
print "Coefficients: " + str(clf.coef_)
```

Based on the above results, we can see that we are decently close to the answer in this case, or at least to know generally when the car is at WOT.  WOT is determined when the APP is at 100% (or close to it) for the duration of the log.  Something to notice here is the *dramatic increase* in MSE.  Because of our unscaled values, we do see a dramatic increase in MSE, but our R^2 value is holding true.  R^2 is also telling us that the model is *effectively explaning the variance seen in the predictions.*

Now, let's take the predicted values and make a guideline to test the manual coefficient calculation for use in Excel or similar software.

##Testing the model

```python
pred_data = pd.DataFrame(clf.predict(lm_data), columns=["Predicted Accel"])
```

Remember the general model for Linear regression: 

$Y = \beta_{0} + \beta_{1}x_{1}+ \beta_{2}x_{2}+ \epsilon$ 

for as many features of x in your model.  With the coefficients above, we have:  

$\beta_{0}$: N/A (We used the non-intercept option on the Linear Regression model)  
$\beta_{1}$: -0.21440671  
$\beta_{2}$: 0.18938429    
$\beta_{3}$: 1.18471672  
$\beta_{4}$: 0.33772201

So in this case, we are using the following equation:

$Y =\beta_{1}$(AFR)$+ \beta_{2}$(Intake Valve Advance)$+ \beta_{3}$(Throttle)$ + \beta_{4}$(Wastegate)

Finally,

**Y = (-0.21440671)x(AFR) + (0.18938429)x(Intake Valve Advance) + (1.18471672)x(Throttle) + (0.33772201)x(Wastegate)**



Now let's take that equation, and run the original data values through it, and see if we get what the prediction function got as a result.

```python
MLR_test = []
for data in lm_data.values:
    MLR_test.append(clf.coef_[0]*(data[0]) + (clf.coef_[1]*(data[1])) + (clf.coef_[2] * (data[2])) + (clf.coef_[3] * (data[3])))
plt.title("Accelerator pedal position (generated)", fontsize=12)
plt.ylabel("Time", fontsize=12)
plt.xlabel("APP", fontsize=12)
plt.plot(MLR_test)
plt.plot(pred_data)
plt.show()
```

##Analysis of the results

So it seems that our model is pretty good, even with only 4 features in the model.  But have we really made a good model?  Let's see what our results tell us.

Our final model is:
Y(APP) = (-0.21440671)x(AFR) + (0.18938429)x(Intake Valve Advance) + (1.18471672)x(Throttle) + (0.33772201)x(Wastegate)

Why do these variables remain in the model?  

- AFR is able to determine when the car is off throttle easily when it runs lean (> 14.7), and can determine decel and accel conditions of the engine
- Intake valve advance is tied to the increase in load of the car
- Throttle nearly matches what the APP is in the car, especially from old cable-driven cars.  Note the high coefficient that is greater than 1
- Wastegate tells us when the car is in high-load acceleration, as additional turbo pressure is desired.

##Is this model prone to car-to-car error?

Remember the old ruling: *"All models are wrong, but some are useful"* by George Box.  Our model has car-to-car variations because the greatest predictors are also the tunable characteristics of the car.  As such, the model can only be assumed for *same configuration* of the engine.  

Since one may change throttle mappings, wastegate tables, and the majority of the predictors, care must be taken to use the model as only a guideline, and not an exact value.  As such, it should only be used to determine the *range* of values that should be considered for further analysis in a parameter log.  

##Are there better models possible?

Yes—this is only one way of tackling the problem.  If features that are not associated with the tunable engine parameters get used in the model, a higher scoring model with less bias may be possible.
